# Modelo de Malthus

Malthus en su <a href="https://es.wikipedia.org/wiki/Ensayo_sobre_el_principio_de_la_poblaci%C3%B3n"> ensayo sobre el principio de la población (1798)</a> interpreta los datos de crecimiento de la población en EEUU en el siglo XVIII reflejando los siguientes puntos: 

* La población sigue una progresión geométrica (crecimiento exponencial).
* Los recursos siguen una progresión aritmética (crecimiento lineal)

    *No veo manera por la que el hombre pueda eludir el peso de esta ley, que abarca y penetra toda la naturaleza animada. Ninguna pretendida igualdad, ninguna reglamentación agraria, por radical que sea, podrá eliminar, durante un siglo siquiera, la presión de esta ley, que aparece, pues, como decididamente opuesta a la posible existencia de una sociedad cuyos miembros puedan todos tener una vida de reposo, felicidad y relativa holganza y no sientan ansiedad ante la dificultad de proveerse de los medios de subsistencia que necesitan ellos y sus familias.*

Esta ley se representa matemáticamente en forma de diferencias como:

$$ \frac{\Delta x}{\Delta t} = \alpha x $$

Que se puede integrar fácilmente viendo que resulta en una forma exponencial:

$$ x(t) = x_0 e^{\alpha t}$$

```python
from matplotlib import pyplot as plt
import numpy as np
plt.xkcd()
```

```python
alpha=0.5
y=[1]
for i in range(10):
    y.append(y[-1]+y[-1]*alpha)
ax=plt.figure(figsize=(10, 6)).add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(y, label='poblacion')
ax.plot(list(map(lambda x: x*3+4, range(11))), label='recursos')
plt.annotate(
    'Catastrofe malthusiana',
    xy=(8, 30), arrowprops=dict(arrowstyle='->'), xytext=(4, 40))
plt.legend()
plt.title('Ley de Malthus')
```

Estas predicciones llevarían irremediablemente a un momento en el que el crecimiento poblacional supera los recursos disponibles, lo que se conoce como una **catástrofe malthusiana**. 

Resultan interesantes los estudios de algunos autores como <a href="https://es.wikipedia.org/wiki/Colapso_(libro)">Jared Diamond</a> que relacionan con las teorías malthusianas catástrofes de determinadas sociedades como el *genocidio de Ruanda* o el colapso poblacional de la *isla de Pascua*.

No obstante, por suerte y en general hay dos factores que llevan a que los pronósticos de Malthus no se cumplan:
* La mayoría de las sociedades acomodadas tienden en un momento determinado hacia un descenso de la natalidad, es lo que se denomina <a href="https://es.wikipedia.org/wiki/Transici%C3%B3n_demogr%C3%A1fica">transición demográfica</a>.
* Los avances tecnológicos llevan a superar ese crecimiento lineal de los recursos disponibles.

# Ecuación logística

Este modelo tan sencillo de crecimiento exponencial funciona muy bien para predecir el crecimiento de colonias bacterianas en las primeras fases, hasta que la colonia comienza a ocupar todo el caldo de cultivo.

¿Qué sucede cuando la población se acerca a esos límites en resursos? Resulta intuitivo pensar que si la población vive en un espacio limitado la concurrencia entre individuos será uno de los principales moderadores de crecimiento. La competencia puede desencadenar luchas entre individuos, la mayor cercanía favorecerá la transmisión de enfermedades... En definitiva hay muchos factores que claramente dependen de la frecuencia entre contactos de los individuos que se modelaría en una primera aproximación con $x^2$.

Podríamos, por tanto, extender nuestra ecuación de la dinámica de la población con un nuevo término tal que:

$$ \Delta x = \alpha x -\beta x^2 $$

Es la ecuación de **Verhulst** o ecuación logística.

```python
alpha=1.5
beta=0.005
y_l=[1]
y=[1]
for i in range(200):
    y.append(y[-1]*alpha)
    y_l.append(y_l[-1]*alpha-y_l[-1]**2*beta)
ax=plt.figure(figsize=(10, 6)).add_subplot(1, 1, 1)
ax.set(ylim=(0,150))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(y[:20], label='Crecimiento exponencial')
ax.plot(y_l[:20], label='Crecimiento logistico')
plt.legend()
```

```python
alpha=0.25
beta=0.05
y_l=[0.1]
y=[0.1]
for i in range(40):
    y.append(y[-1] + y[-1]*alpha)
    y_l.append(y_l[-1] + y_l[-1]*alpha-y_l[-1]**2*beta)
ax=plt.figure(figsize=(10, 6)).add_subplot(1, 1, 1)
ax.set(ylim=(0,alpha/beta*1.1))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(y, label='Crecimiento exponencial')
ax.plot(y_l, label='Crecimiento logistico')
ax.plot([0,40],[alpha/beta, alpha/beta], linestyle='--', label='alpha/beta')
plt.legend()
```

```python
alpha=0.25
beta=0.05
y_l=[0.1]
dy=[alpha]
for i in range(40):
    y_l.append(y_l[-1] + y_l[-1]*alpha-y_l[-1]**2*beta)
    dy.append(y_l[-1]*alpha-y_l[-1]**2*beta)
ax=plt.figure(figsize=(10, 6)).add_subplot(1, 1, 1)
ax.set(ylim=(0,alpha/beta*1.2))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax2=ax.twinx()
ax.plot(y_l, label='Crecimiento logistico', color='black')
ax.plot([0,40],[alpha/beta, alpha/beta], linestyle='--', color='black', label='alpha/beta')
ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')
ax2.set(yticks=[])
ax2.set(ylim=(0,max(dy)*1.5))
ax2.plot(dy[1:], label='Ritmo de crecimiento', linestyle='--')
ax.annotate(
    'Cambio de pendiente',
    xy=(15, 4), arrowprops=dict(arrowstyle='->'), xytext=(0, 4.5))
ax.annotate(
    'Cambio de pendiente',
    xy=(15, 2), arrowprops=dict(arrowstyle='->'), xytext=(0, 4.5))
plt.legend()
```

# Modelo SIR

Una extensión sencilla de este modelo aplicado a epidemias es el modelo SIR que distingue entre individuos susceptibles ($S$), infectados ($I$) y recuperados ($R$). 

* $ \Delta S = - \beta \frac{S  I}{n_T}$
* $ \Delta I =  \beta \frac{S  I}{n_T} - \gamma I$
* $ \Delta R = \gamma  I$

Donde $n_T$ es la población total ($S+I+R$), $\beta$ regula la probabilidad de que un individuo susceptible pase a infectarse y $\gamma$ regula la probabilidad de que un infectado se recupere.

Al analizar la evolución de los infectados podemos observar que tiene la misma forma que la ecuación logística.

Con la misma lógica, y en pos de simular lo más exactamente posible los casos que nos ocupan, vamos a introducir un término adicional análogo al de recuperados para tener en cuenta las muertes:

* $ \Delta S = - \beta \frac{S  I}{n_T}$
* $ \Delta I =  \beta \frac{S  I}{n_T} - (\gamma_r+\gamma_d) I$
* $ \Delta R = \gamma_r  I$
* $ \Delta D = \gamma_d  I$

Con esta modificación lo implementamos:

```python
import pandas
class SIR_model(object):
    def __init__(self, S0, I0, R0, D0, beta, gamma, gamma_d, delay=10):
        self._history = {}
        self._history['S']=[S0]
        self._history['I']=[I0]
        self._history['R']=[R0]
        self._history['D']=[D0]
        self.beta = beta
        self.gamma = gamma
        self.gamma_d = gamma_d
        self.nT = S0 + R0 + I0 + D0
        self.delay = delay
    @property
    def S(self):
        return self._history['S'][-1]
    @property
    def I(self):
        return self._history['I'][-1]
    @property
    def I_delayed(self):
        if len(self._history['I']) <= self.delay:
            return self._history['I'][0]
        else:
            return self._history['I'][-(self.delay)+1]
    @property
    def R(self):
        return self._history['R'][-1]
    @property
    def D(self):
        return self._history['D'][-1]
    def step(self):
        S_1 = self.S - self.beta * self.S * self.I / self.nT
        I_1 = self.I + self.beta * self.S * self.I / self.nT - (self.gamma + self.gamma_d ) * self.I
        R_1 = self.R + self.gamma * self.I_delayed
        D_1 = self.D + self.gamma_d * self.I_delayed
        self._history['S'].append(S_1)
        self._history['I'].append(I_1)
        self._history['R'].append(R_1)
        self._history['D'].append(D_1)
    def steps(self, n):
        for i in range(n):
            self.step()
    @property
    def history(self):
        return pandas.DataFrame(self._history)
```

Lanzando el modelo con unos parámetros arbitrarios:

```python
m=SIR_model(9999,1,0,0,0.5,0.02,0.01)
```

```python
m.steps(100)
h=m.history
h.plot()
```

La suma de individuos infectados, recuperados y muertos nos dará el total de afectados, que sigue una evolución de tipo logístico:

```python
h['c']=h.I+h.R+h.D
h.c.plot(title='Confirmados')
```

# Datos reales del caso Covid-19

Vamos a partir de los datos publicados por **Johns Hopkins University Center for Systems Science and Engineering (JHU CSSE)**.

https://github.com/CSSEGISandData/COVID-19

```python
import pandas
Confirmed = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
Deaths = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
Recovered = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv')
```

```python
PROVINCES = Recovered[['Province/State', 'Country/Region', 'Lat', 'Long']].fillna('')
PROVINCES['Location']= PROVINCES['Country/Region'] + ' ' + PROVINCES['Province/State']
PROVINCES['Location'] = PROVINCES['Location'].apply(lambda x: x.strip())
PROVINCES
dates = Confirmed.columns[4:]
#  location, long, lat, date, c, d, r
data={'location':[],'date':[],'long':[],'lat':[],'C':[],'D':[],'R':[], 'days_from_first_c': []}
days_from_first={}
for i in PROVINCES.index:
    days_from_first[PROVINCES['Location'][i]] = 0
    for d in dates:          
        data['location'].append(PROVINCES['Location'][i])
        data['lat'].append(PROVINCES['Lat'][i])
        data['long'].append(PROVINCES['Long'][i])
        data['date'].append(d)
        data['C'].append(Confirmed[d][i])
        data['R'].append(Recovered[d][i])
        data['D'].append(Deaths[d][i])
        data['days_from_first_c'].append(days_from_first.get(PROVINCES['Location'][i], 0))
        if Confirmed[d][i] > 0:
            days_from_first[PROVINCES['Location'][i]] += 1
df = pandas.DataFrame.from_dict(data)
df['date']=pandas.to_datetime(df['date'])
```

Extrayendo los datos para la región de Hubei:

```python
hubei = df[df['location']== 'China Hubei'].set_index('days_from_first_c')
hubei['I']=hubei.C-hubei.R-hubei.D
#hubei['I'] = hubei['I'].rolling(3, win_type='triang').mean()
hubei[['I', 'C', 'R', 'D']].plot(title='China Hubei')
```

Podemos intentar visualizar las constantes que estamos teniendo en cuenta en el modelo:

```python
nt=hubei.C.max()
hubei['S']=nt-hubei['I']
hubei['beta']=(hubei.C.diff(2)*nt/(hubei.S*hubei.I))
hubei['beta'].plot(title='Beta')
```

```python
hubei['gamma_r']=(hubei.R.diff(2)/hubei.I)
hubei['gamma_r'].plot(title='Gamma Recuperados')
```

```python
hubei['gamma_d']=(hubei.D.diff(2)/hubei.I)
hubei['gamma_d'].plot(title='Gamma Muertos')
```

```python
beta=hubei['beta'].head(20).mean()
gamma_d=hubei['gamma_d'].median()
gamma_r=hubei['gamma_r'].median()
(hubei.C.max(),hubei.I[0],hubei.R[0],hubei.D[0],beta,gamma_r,gamma_d, 5)
```

```python
def eval(i, beta, gamma_r, gamma_d, delay):
    m=SIR_model(hubei.C.max(),hubei.I[0],hubei.R[0],hubei.R[0],beta,gamma_r,gamma_d, delay)
    m.steps(60)
    h=m.history
    return (h[i]-hubei[i]).abs().sum()
```

```python
def dummy_fit(beta, gamma_r, gamma_d, delay): 
    delta_g = 0.00001
    delta_d = -1
    delta_b = 0.0001
    if eval('I', beta, gamma_r, gamma_d, delay) < eval('I', beta+delta_b, gamma_r, gamma_d, delay):
        delta_b*=-1
    while eval('I', beta, gamma_r, gamma_d, delay) > eval('I', beta+delta_b, gamma_r, gamma_d, delay):
        beta+=delta_b

    if eval('R', beta, gamma_r, gamma_d, delay) < eval('R', beta, gamma_r, gamma_d, delay+delta_d):
        delta_d*=-1
    while eval('R', beta, gamma_r, gamma_d, delay) > eval('R', beta, gamma_r, gamma_d, delay+delta_d):
        delay+=delta_d

    if eval('R', beta, gamma_r, gamma_d, delay) < eval('R', beta, gamma_r+delta_g, gamma_d, delay):
        delta_g*=-1
    while eval('R', beta, gamma_r, gamma_d, delay) > eval('R', beta, gamma_r+delta_g, gamma_d, delay):
        gamma_r+=delta_g

    if eval('R', beta, gamma_r, gamma_d, delay) < eval('R', beta, gamma_r, gamma_d, delay+delta_d):
        delta_d*=-1
    while eval('R', beta, gamma_r, gamma_d, delay) > eval('R', beta, gamma_r, gamma_d, delay+delta_d):
        delay+=delta_d
        
    if eval('D', beta, gamma_r, gamma_d, delay) < eval('D', beta, gamma_r, gamma_d+delta_g, delay):
        delta_g*=-1
    while eval('D', beta, gamma_r, gamma_d, delay) > eval('D', beta, gamma_r, gamma_d+delta_g, delay):
        gamma_d+=delta_g

    return (beta, gamma_r, gamma_d, delay)
#(beta, gamma_r, gamma_d, delay) = ( 0.4478821168534803,0.06472708237984669,0.0040919369044244095,5)
for i in range(20):
    (_beta, _gamma_r, _gamma_d, _delay) = dummy_fit(beta, gamma_r, gamma_d, delay)
    if (_beta, _gamma_r, _gamma_d, _delay) == (beta, gamma_r, gamma_d, delay):
        break
    (beta, gamma_r, gamma_d, delay) = (_beta, _gamma_r, _gamma_d, _delay)    
    print(beta, gamma_r, gamma_d, delay)
    
```
